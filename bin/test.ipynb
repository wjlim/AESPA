```python
import pandas as pd
```

```python
order_info = '../assets/OrderInfo.txt'
sample_info = '../assets/samplesheet.csv'

order_df = pd.read_table(order_info)
sample_df = pd.read_csv(sample_info)
```

```python
sample_df_indexed = sample_df.set_index('SampleID')
order_df_indexed = order_df.set_index('SampleID')
common_columns = order_df_indexed.columns.intersection(sample_df_indexed.columns)
sample_df_indexed = sample_df_indexed.drop(columns=common_columns)
merged_df = pd.merge(order_df_indexed, sample_df_indexed, left_index=True, right_index=True, how='right')
merged_df.reset_index(inplace=True)
```
